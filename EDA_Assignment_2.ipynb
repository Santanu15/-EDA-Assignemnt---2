**Amazon Sales Data**


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('amazon.csv')
```

```python
df
```

```python
# What is the average rating for each product category
```

```python
df.head()
```

```python
df['rating']=pd.to_numeric(df['rating'],errors='coerce')

avg_rating=df.groupby('category')['rating'].mean().reset_index()

avg_rating.columns=['category','avg_rating']

avg_rating.sort_values(by='avg_rating',ascending=False,inplace=True)

avg_rating

```

```python
# What are the top rating_count products by category
```

```python
df.head()
```

```python
if df['rating_count'].dtype=='object':
  df['rating_count']=df['rating_count'].str.replace(',','').astype(float)

def top_product(group):
  return group.sort_values(by='rating_count',ascending=False).head()

top_product_by_category=df.groupby('category').apply(top_product).reset_index(drop=True)

top_product_by_category=top_product_by_category[['category','product_name','rating_count','rating']]

top_product_by_category
```

```python
# What is the distribution of discounted prices vs. actual prices
```

```python
df.head()
```

```python
if df['discounted_price'].dtype == 'object':
  df['discounted_price']=df['discounted_price'].str.replace('₹','').str.replace(',','').astype(float)

if df['actual_price'].dtype == 'object':
  df['actual_price']=df['actual_price'].str.replace('₹','').str.replace(',','').astype(float)

plt.figure(figsize=(6,5))

sns.histplot(df['discounted_price'],bins=50,color='blue',kde=True,label='Discounted_price')

sns.histplot(df['actual_price'],bins=50,color='red',kde=True,label='Actual_price')

plt.title('Distribution of Discounted Price vs. Actual price')

plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend()
plt.show()

```

```python
# How does the average discount percentage vary across categories
```

```python
df.head()
```

```python
if df['discount_percentage'].dtype == 'object':
  df['discount_percentage']=df['discount_percentage'].str.replace('%','').astype(float)

avg_discount=df.groupby('category')['discount_percentage'].mean().reset_index()

avg_discount=avg_discount.sort_values(by='discount_percentage',ascending=False)

plt.figure(figsize=(10,8))

sns.barplot(data=avg_discount,x='discount_percentage',y='category',palette='viridis')

plt.title('Average Discount Percentage by Category')

plt.xlabel('Average Discount Percentage')
plt.ylabel('Category')
plt.show()
```

```python
# What are the most popular product names
```

```python
df.head()
```

```python
if df['rating_count'].dtype == 'object':
  df['rating_count']=df['rating_count'].str.replace('.','').astype(float)

top_product=df[['product_name','rating_count']].sort_values(by='rating_count',ascending=False).head(10)

plt.figure(figsize=(10,8))

sns.barplot(data=top_product,x='rating_count',y='product_name',palette='Spectral')

plt.title(' Top 10 Most Popular Product Names')
plt.xlabel('Rating Count')
plt.ylabel('Product Name')
plt.show()
```

```python
# What are the most popular product keywords
```

```python
df.head()
```

```python
import re
from collections import Counter
```

```python
def token(text):
  text=re.sub(r'[^\w\s]','',text)
  words=text.lower()
  return words


keywords=df['product_name'].dropna().apply(token).sum()

key_counts=Counter(keywords)

keyword_df=pd.DataFrame(key_counts.items(),columns=['keyword','count']).sort_values(by='count',ascending=False).head(10)

plt.figure(figsize=(6,5))

sns.barplot(data=keyword_df,x='count',y='keyword',palette='viridis')

plt.title('Top 10 Most Popular Product Keywords')
plt.xlabel('Count')
plt.ylabel('Keyword')
plt.show()
```
